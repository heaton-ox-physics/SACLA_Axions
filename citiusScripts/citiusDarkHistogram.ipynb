```python
runList = []
for i in range(255550,255587+1):
    runList.append(i)
fileName = "citiusData/citiusHistogram_BL2_r" + str(runList[0]) + ".npy"
histogram = np.load(fileName)
bins = []
counts = []
for i in range(np.shape(histogram)[0]):
    bins.append(histogram[i][0])
    counts.append(histogram[i][1])
    
totalCounts = np.zeros(np.shape(counts))
numberOfRuns = 0
for run in runList:
    try:
        fileName = "citiusData/citiusHistogram_BL2_r" + str(run) + ".npy"
        histogram = np.load(fileName)

    except Exception as e:
        print(e)
        break
    counts = []
    individualBins = []
    
    for i in range(np.shape(histogram)[0]):
        individualBins.append(histogram[i][0])
        counts.append(histogram[i][1])
    totalCounts = counts + totalCounts
    # plt.title(run)
    # plt.bar(individualBins,counts)
    # plt.yscale("log")
    # plt.xlim(0,10)
    # plt.show()
    numberOfRuns = numberOfRuns + 1
        
plt.bar(bins,totalCounts)
plt.yscale("log")
tileString = "All runs: " + str(round(numberOfRuns * 3000 / 30 / 60 / 60,2)) + " hours"
plt.title(tileString,fontsize=15,fontfamily="serif")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Pixel value",fontsize=15,fontfamily="serif")
plt.ylabel("Number of counts",fontsize=15,fontfamily="serif")
```
